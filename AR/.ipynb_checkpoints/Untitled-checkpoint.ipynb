```python
d1602.describe()
```

```python
#iz podatkov vzamem le osebne avtomobile (vozila z oznako M1)
dataOA = data[data['J-Kategorija in vrsta vozila (oznaka)'] == "M1"]
dataOA
```

```python
#ker bom primerjal po starosti iz podatkov vzamem le vnose, kjer je znana starost
dataOAS = dataOA[dataOA['C-Starost uporabnika vozila']>0]
dataOAS
```

```python
#starost spremenim v kategorije
dataOAS['C-Starost uporabnika vozila'] = pd.cut(dataOAS['C-Starost uporabnika vozila'], 10, retbins=True, precision=0)[0]
dataOAS
```

```python
dataOAS['P.1.2-Nazivna moc']
```

```python
ageHP = dict()
for a in dataOAS['C-Starost uporabnika vozila']:
    if(a not in ageHP):
        ageHP[a] = [dataOAS['P.1.2-Nazivna moc'][dataOAS['C-Starost uporabnika vozila']==a].values]
    #else:
    #    ageHP[a].append(dataOAS['P.1.2-Nazivna moc'][dataOAS['C-Starost uporabnika vozila']==a].values[0])
    #df.loc[df['Letters'] == 'C', 'Letters'].values[0]
    #print(dataOAS['P.1.2-Nazivna moc'][dataOAS['C-Starost uporabnika vozila']==a].values[0])
#ageHP
```

```python
avgHP = []
for ag in ageHP:
    avgHP.append(sum(ageHP[ag][0]) / len(ageHP[ag][0]))
avgHP
```

```python
len(ageHP.keys())
```

## Povprečna nazivna moč osebnih avtomobilov glede na starostno skupino ##

```python
starost = ageHP.keys()

x = range(len(starost))
y = avgHP

plt.plot(x, y)
plt.ylim(ymax = 100, ymin = 10)
plt.xticks(range(len(starost)), sorted(starost))
plt.xticks(rotation=90)
plt.ylabel('Povprečna nazivna moč')
plt.xlabel('Starostna skupina')
for i,j in zip(x,y):
    plt.annotate(str(np.around(j, 2)),xy=(i,j+2))
plt.savefig('starost_nazivnamoc.png', bbox_inches='tight')
plt.show()
```
